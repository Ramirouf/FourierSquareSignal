Desarrollo de código en Python para representar los primeros siete armónicos de una señal cuadrada.

Se utilizaron las librerías numpy para crear la señal y matplotlib para graficarla.

```python
import numpy as np
import matplotlib.pyplot as plt

T = 2  # Período de la señal cuadrada
t = np.linspace(0, 2 * T, 1000)  # Intervalo de tiempo
f_t = np.zeros_like(t)

# Suma de los primeros siete términos de la Serie de Fourier
for n in range(1, 8):
    f_t += (1 / (2 * n - 1)) * np.sin((2 * n - 1) * np.pi * t / T)

# Escalar la señal para que esté en el rango [0, 1]
f_t = 0.5 + 0.5 * f_t

plt.plot(t, f_t)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Aproximación de la Señal Cuadrada con la Serie de Fourier (7 términos)')
plt.grid(True)
plt.ylim(0, 1)  # Asegura que la señal esté en el rango [0, 1]
plt.show()
```
